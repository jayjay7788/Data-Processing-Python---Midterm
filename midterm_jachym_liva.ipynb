```python
## data import

zip_file_path = 'data_midterm.zip'
dfs = list()

with zipfile.ZipFile(zip_file_path, 'r') as z:
    print(z.namelist()) 
    
    csv_filenames = ['data_ws2025/MHK.csv', 'data_ws2025/KR.csv', 'data_ws2025/HST.csv', 'data_ws2025/LOW.csv', 'data_ws2025/RTX.csv']
    
    for file in csv_filenames:
        with z.open(file) as f:
            data = pd.read_csv(f, parse_dates=["Date"], date_format="%Y-%m-%d")
            data['Ticker'] = f'{file}'
            dfs.append(data)
    

returns = pd.concat(dfs, ignore_index=True)
print(returns.tail())

type(returns.Date[0])
```

```python
## 1
(returns.High == returns.Low).any()

## there is no such company
```

```python
## 2 
companies = returns.Ticker.unique()

max_price = []
min_price = []

for company in companies:
    maxp = returns[returns.Ticker == company].High.max() 
    minp = returns[returns.Ticker == company].Low.min() 
    max_price.append(maxp)
    min_price.append(minp)

prices = pd.DataFrame([companies, max_price, min_price])
prices
```

```python
## 3

returns['log_close'] = np.log(returns.Close) 
#returns.head()

summary = pd.DataFrame()

for company in companies:
    summary[f'company'] = returns[returns.Ticker == company].aggregate([np.mean, np.std, np.min, np.max])
```
